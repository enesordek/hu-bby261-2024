```python
import pandas as pd
import matplotlib.pyplot as plt

veriSeti = pd.read_csv('ogrenciVeri.csv')
ogrenciCinsiyet = veriSeti['Gender']
basarıDurumu = veriSeti['Success']

plt.figure(figsize=(20, 7))  # Grafiğin boyutunu ayarlayın
plt.bar(ogrenciCinsiyet, basarıDurumu)
plt.xlabel('ÖĞRENCİ CİNSİYETLERİ')
plt.ylabel('BAŞARI DURUMU')
plt.title('Cinsiyete göre öğrenci Başarısı')
plt.xticks(rotation=90)  # Ülke isimlerinin dikey yazılması
plt.show()
```
